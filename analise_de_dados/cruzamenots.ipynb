```python
df = pd.read_csv("dados/DadosBrutos_Palmeiras_x_Bragantino.csv")
df.head(20)
```

```python
# OBSERVAR SOMENTE A BOLA

filtro_bola = df["OBJETO_OBSERVADO"] == "bola"
df_bola = df[filtro_bola]
df_bola
```

```python
# TIRAR A BOLA PARADA

filtro_movimento = df["VELOCIDADE"] > 1
df_bola = df_bola[filtro_movimento]
df_bola
```

```python
filtro_pal = df_bola["TIME_POSSE"] == "PAL"
df_bola = df_bola[filtro_pal]
df_bola
```

```python
filtro_bola_em_jogo = df["BOLA_EM_JOGO"] == 1
df_bola = df_bola[filtro_bola_em_jogo]
```

```python
# VERIFICAR SE A BOLA ESTA NA AREA PROPICIA PARA SER CRUZAMENTO

# filtro_cruz = df_bola["X"] > 90
# df_bola = df_bola[filtro_cruz]
# df_bola
```

```python
# VERIFICAR SE A BOLA ESTA NA AREA INIMIGA NO PRIMEIRO TEMPO

x = 104.5
y1 = 33
y2 = 51

filtro_1_tempo = df_bola["TEMPO_JOGO"] == 1
df_bola = df_bola[filtro_1_tempo]

df_bola["DENTRO_AREA_DIREITA"] = 0

area_direita = (df_bola["X"] > x) & (df_bola["Y"] > y1) & (df_bola["Y"] < y2)
df_bola.loc[area_direita, "DENTRO_AREA_DIREITA"] = 1
df_bola = df_bola.reset_index()
df_bola
```

```python
frames = []

for i in range(len(df_bola)):
    if df_bola.iloc[i]["DENTRO_AREA_DIREITA"]:
        frames.append(df_bola.iloc[i]["FRAME"])

print(frames)
```

```python
count = 0
frames_dentro_area_direita = []
for i in range(len(frames)-1):
    if not (frames[i+1] - frames[i]) > 50 :
        count += 1
    else:
        frames_dentro_area_direita.append(frames[i-count])
        count = 0
    
print(len(frames_dentro_area_direita))
print((frames_dentro_area_direita))

```
